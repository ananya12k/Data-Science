```python
import pandas as pd
import random as ran
import numpy as np
```

```python
# Loading Data from csv file 
details=pd.read_csv('Details.csv')
print(details)
```

```python
# Loading Data from csv file 
placement=pd.read_csv('Placement.csv')
print(placement)
```

```python
merged_data = pd.merge(details, placement, on='Roll No.')
print(merged_data)
```

```python
placed_stu=merged_data[merged_data['Placed']=='Yes'].copy()
placed_stu
```

```python
# Add salary according to CPI
for i in  placed_stu.index:
    CPI=placed_stu.loc[i].at['CPI']
    if CPI >= 8:
        placed_stu.loc[i, 'Salary'] = ran.randint(65000, 70000)
    elif CPI >= 7 and CPI < 8:
        placed_stu.loc[i, 'Salary'] = ran.randint(60000, 65000)
    else:
        placed_stu.loc[i, 'Salary'] = ran.randint(55000, 60000)
```

```python
placed_stu
```

```python
placed_stu['CPI']=placed_stu['CPI'].fillna(0)
cpi=placed_stu['CPI'].to_numpy()
placed_stu['CPI'] = placed_stu['CPI'].replace(to_replace='0.00', value=np.mean(cpi))
placed_stu
```

```python
from datetime import datetime, timedelta

event_date = datetime(year=2024, month=1, day=9)  # Modify based on your reference date

# Calculate birth years directly from the "Age" column
placed_stu["birth_year"] = event_date.year - placed_stu["Age"]

# Generate random days within each birth year
placed_stu["birth_date"] = placed_stu["birth_year"].apply(
    lambda year: datetime(year=year, month=random.randint(1, 12), day=random.randint(1, 28))
)
```

```python
placed_stu
```

```python
# Filter
# Create list whose cpi is above 8
excellent=placed_stu[placed_stu['CPI']>=8].copy()
excellent
```
