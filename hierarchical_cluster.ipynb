```python
# DBScan Algorithm
# Aspects of the algorithm :
# 1. Core points
# 2. Border points
# 3. Noise points

# Core points : Start with a point and check if there are atleast min_samples within a radius epsilon of the point. If yes, then the point is a core point and all the points within the radius are part of the same cluster. If no, then the point is a noise point. Grouping of points is done by the algorithm itself.

# Border points : Points that are within the radius of a core point but do not have min_samples within the radius of epsilon. These points are part of the cluster but are not core points.

# Noise points : Points that are neither core points nor border points.

# Why DBScan?
# 1. It can find arbitrarily shaped clusters
# 2. It can handle noise
# 3. It can handle outliers
# 4. It can handle clusters of different shapes and sizes
# 5. It can handle clusters of different densities

# How many clusters are there?
#  Has global optimum for the number of clusters and the number of clusters is not required to be specified beforehand. 
# In case of KMeans, the number of clusters is required to be specified beforehand.

#  Time complexity is O(nlogn) for DBScan

# What happenned in KMeans that DBScan solves? Give an example
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import dbscan
X, y_true = make_blobs(n_samples=200, centers=6, cluster_std=0.6)
plt.scatter(X[:,0],X[:,1], s=15)
```

```python
dbs = dbscan(X)
plt.scatter(X[:, 0], X[:, 1], c=dbs[1], s=15, cmap='viridis')
```

```python
# Perform DBScan on the iris dataset for 2 features
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
iris = datasets.load_iris()
print(iris.feature_names)

# plot the iris dataset
plt.scatter(iris.data[:,0], iris.data[:,1], s=15)
```

```python
# create a dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head())
#  Select the first 2 features
X = df.iloc[:, [0, 1]].values
#  Perform DBScan
dbs = DBSCAN(eps=0.5, min_samples=5).fit(X)

#  Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=dbs.labels_, s=15, cmap='viridis')
plt.show()

```
