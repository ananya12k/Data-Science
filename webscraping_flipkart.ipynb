```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# take a website to perform webscraping 
url= 'https://www.flipkart.com/search?q=fiction+books+english&sid=bks&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_7_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_7_na_na_na&as-pos=2&as-type=RECENT&suggestionId=fiction+books+english%7CBooks&requestId=4e14c60f-8aaf-40be-b83a-7ba38fcf4d34&as-searchtext=fiction%20books%20english'
```

```python
# Flipkart url taken
html=requests.get(url)
s=BeautifulSoup(html.content,'html.parser')
print(s.prettify())
```

```python
# Finding container
containers=s.find(id='container')

# the page has a lot of cards so find the main container by class name
main_container=containers.find_all('div',class_='_1YokD2._3Mn1Gg')
# we will store the details of the books in a dataframe 
# we will use pandas library to store the details of the books in a dataframe

df=pd.DataFrame(columns=['Title','Price','Author_Language','Image'])
for container in main_container:
    title=container.find('a',class_='s1Q9rs').title
    price=container.find('div',class_='_30jeq3').text
    rating=container.find('div',class_='_3LWZlK"').text
    language_and_author=container.find('div',class_='_3Djpdu').text
    image=container.find('img',class_='_396cs4')['src']
    df=df.append({'Title':title,'Price':price,'Author_Language':language_and_author,'Image':image},ignore_index=True)


print(df)

```
